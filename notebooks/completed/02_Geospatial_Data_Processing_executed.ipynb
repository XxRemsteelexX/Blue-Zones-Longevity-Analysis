Processing geographic and spatial data for Blue Zones analysis.


```python
# Setup
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load test data
df = pd.read_csv('../outputs/test_data_with_gravity.csv')
print("üìä Data loaded successfully!")
print(f"Shape: {df.shape}")
df.head()
```

```python
# Basic visualization
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(df['longitude'], df['latitude'], c=df['effective_gravity'], cmap='viridis')
plt.colorbar(label='Gravity (m/s¬≤)')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Global Gravity Distribution')

plt.subplot(1, 2, 2)
blue_zones = df[df['is_blue_zone'] == 1]
others = df[df['is_blue_zone'] == 0]

plt.scatter(others['effective_gravity'], others['life_expectancy'], alpha=0.6, label='Other Regions')
plt.scatter(blue_zones['effective_gravity'], blue_zones['life_expectancy'], 
           color='red', s=100, label='Blue Zones', marker='*')
plt.xlabel('Effective Gravity (m/s¬≤)')
plt.ylabel('Life Expectancy')
plt.title('Gravity vs Longevity')
plt.legend()

plt.tight_layout()
plt.show()

print("üéâ Geospatial analysis complete!")
```
